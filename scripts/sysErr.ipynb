#Uncertainty in difference spectrum

We have two versions of the difference spectrum. One made from subtracting high and low state spectra, the other made from fitting sinusoids with fixed phase and plotting the amplitude as a function of wavelength. These two methods are vunerable to different systematics; the first method can be corrupted by erratic errors in slit loss correction, the latter method is corrupted by non-sinusoidal variability. Thus, the difference between the two gives a crude estimate of the uncertainty in the difference spectrum.

I begin by loading in the red and blue difference spectra produced by the 'sin method' and merging into one spectrum

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# sin method
xr,yr = np.loadtxt('../keck/amplSpectrum_red.txt').T
xb,yb = np.loadtxt('../keck/amplSpectrum_blue.txt').T
# fix 1e17 factor in flux calibration
yr /= 1e17
yb /= 1e17

# simple method
x,y,_ = np.loadtxt('../keck/differenceSpectrum.txt').T

# get data in right ranges
bmask = (4000 <= xb) & (xb  < 5500)
xb = xb[bmask]; yb = yb[bmask]
rmask = (xr <= 8000)
xr = xr[rmask]; yr = yr[rmask]

# put in one spectrum
x2 = np.concatenate((xb,xr))
y2 = np.concatenate((yb,yr))

print x.min(), x.max()
print x2.min(), x2.max()

# Choose a bunch of new plotting parameter values
newparams = {'axes.labelsize': 8, 'axes.linewidth': 1, 'savefig.dpi': 300, 
             'lines.linewidth': 0.5, 'figure.figsize': (8, 5),
             'figure.subplot.wspace': 0.0, 'figure.subplot.hspace': 0.0,
             'ytick.labelsize': 8, 'xtick.labelsize': 8,
             'ytick.major.pad': 5, 'xtick.major.pad': 5,
             'legend.fontsize': 8, 'legend.frameon': False, 
             'legend.handlelength': 1.5}
# Update the global rcParams dictionary with the new parameter choices
plt.rcParams.update(newparams)

# plot
plt.step(x,y,alpha=0.5,color='g')
plt.step(x2,y2,color='b')
plt.show()
```

We have a slight issue here in that the two spectra have different spectral resolutions, so we'll have to rebin the high-resolution one down to the coarser grid before subtracting the two.

```python
from scipy.stats import binned_statistic 
bin_edges = x2
bin_step = np.mean(x2[1:]-x2[:-1])
bin_edges = np.append(bin_edges,[bin_edges[-1]+bin_step])
bin_edges -= bin_step/4.

#binned version of hi-resolution difference spectrum
y_bin,x_bin,bin_nums = binned_statistic(x,y,'mean',bin_edges)

# difference between two is our uncertainty estimate
delta = np.fabs(y_bin-y2)/2.
```

Now, of course, we don't really want to lose the resolution we have in the higher res version, so let's write a routine to interpolate this uncertainty estimate back up to full resolution

```python
from scipy.interpolate import interp1d
def getY(x,y,xgoal):
    func = interp1d(x,y,kind='linear')
    return func(xgoal)

# interpolate back up to high resolution
delta = np.array([getY(x2,delta,val) for val in x])
```

Now, let's use our high resolution, max-min, version of the difference spectrum, and plot that with this uncertainty.

```python
plt.step(x,y,color='k',lw=0.5)
# plot a gray shaded area that represents +/- 1 sigma uncertainty.
plt.fill_between(x,y-delta,y+delta,color='k',alpha=0.2)
plt.axhline(0.0)
plt.show()
```

For comparison, let's show the uncertainties suggested by the noise "nuisance parameter", $\sigma$, found from the MCMC fit. This is 0.018 mJy *at all wavelengths*.

```python
plt.step(x,y,color='k',lw=0.5)
delta_mcmc = 0.018 # mcmc nuisance parameter
# plot a gray shaded area that represents +/- 1 sigma uncertainty.
plt.fill_between(x,y-delta_mcmc,y+delta_mcmc,color='k',alpha=0.2)
plt.show()
```

You can see that by forcing $\sigma = 0.018$ mJy everywhere, we're downweighting the blue side of the fit compared to the red. I could re-do the MCMC fitting with the fancier estimate of the uncertainty on the difference spectrum. Either for this, or for plotting, we'll need a version of the spectrum with these uncertainties written to disk, so let's do this.
